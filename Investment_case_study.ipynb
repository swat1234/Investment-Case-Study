## Task 1: Reading and Inspection

-  ### Import and read

Import and read the rounds2 and companies database. Store it in a variable called `rounds2` and `companies`.

```python
# ISO8859-1 is used as the encoding scheme

rounds2 = pd.read_csv('rounds2.csv',sep = ',',encoding='ISO8859-1')
companies = pd.read_csv('companies.txt',sep = '\t',encoding='ISO8859-1')
```

 ### As there are lot of of Garbage character, encoding and decoding mechanism is applied

```python
rounds2['company_permalink']=rounds2['company_permalink'].str.encode('utf-8').str.decode('ascii','ignore')
companies['permalink']=companies['permalink'].str.encode('utf-8').str.decode('ascii','ignore')
```

### Converting the values of company_permalink and permalink to lower case

```python
rounds2['company_permalink'] = rounds2['company_permalink'].str.lower()
companies['permalink']=companies['permalink'].str.lower()
```

### Finding if there are any special characters left

```python
print(len(rounds2.loc[~rounds2['company_permalink'].isin(companies['permalink']),:]))
```

## Dropping unnecessary Columns

### For the case study, from rounds2 data, `funding_round_code`,`founded_at` are not useful for analysis
### From companies data, `homepage_url`, `state_code`,`region`,`city`,`founded_at` are not useful for analysis. Therefore dropping all those columns

```python
companies.isnull().sum()
```

```python
rounds2.isnull().sum()
```

```python
rounds2=rounds2.drop(['funding_round_code','funded_at'],axis=1)
```

```python
companies = companies.drop(['homepage_url','state_code','region','city','founded_at'],axis=1)
```

### Dropping rows with null values as they will impact the analysis

```python
#finding the rows with null values for companies

companies.isnull().sum()
```

```python
#finding the rows with null values for rounds2
rounds2.isnull().sum()
```

```python
# Dropping the rows for raised_amount_usd, country_code,category_list

rounds2=rounds2[~np.isnan(rounds2['raised_amount_usd'])]

companies = companies[pd.notnull(companies.country_code)]
companies = companies[pd.notnull(companies.category_list)]
```

### Finding the Unique companies in `companies` sheet After Data Cleaning

```python
len(list(companies['permalink'].unique()))
```

### Finding the Unique companies in `rounds2` sheet after Data Cleaning

### Finding if there any companies in the `rounds2` file which are not  present in `companies`

```python
len(set(rounds2.company_permalink)-set(companies.permalink))
```

### Merging both the rounds2 and companies data and creating a Master DataFrame

```python
master_frame = pd.merge(companies,rounds2,how='inner',left_on="permalink",right_on="company_permalink")
```

```python
master_frame.isnull().sum()
```

# Checkpoint 2.1
### Grouping the data by `funding_round_type` and finding the average  of `raised_amount_usd`

```python
funding_group=master_frame.groupby('funding_round_type')
master_frame.funding_round_type.value_counts()
```

```python

Funding_type =pd.DataFrame(funding_group.raised_amount_usd.mean().sort_values(ascending=False).apply(lambda x: "{:,.2f}".format(x)))
Funding_type["Total_no_of_investments"] = master_frame.funding_round_type.value_counts()
Funding_type.loc[["venture","seed","angel","private_equity"],:]
```

# Checkpoint 3.1

# Checkpoint 4.1
### Splitting the category_list and extracting the first category and storing it in a new column as `primary_sector`

```python
master_frame['category_list']=master_frame['category_list'].str.split('|')
master_frame['primary_sector']= master_frame['category_list'].apply(lambda x:x[0])
```

### Loading the mapping data  and converting the data into a unpivotable data using pd.melt

```python
mapping = pd.read_csv("mapping.csv")
```

```python
mapping = pd.melt(mapping, id_vars=['category_list'], value_vars=['Automotive & Sports', 'Blanks',"Cleantech / Semiconductors",'Entertainment','Health','Manufacturing','News, Search and Messaging','Others','Social, Finance, Analytics, Advertising'],value_name='Value')
```

```python
mapping=mapping.loc[mapping.Value == 1]
```

### Inspecting the category list in the mapping data and impute the data accordingly

```python
mapping=mapping.rename(columns={'variable':'main_sector'})
```

```python
master_frame['primary_sector']=master_frame['primary_sector'].str.replace('na','0')
master_frame['primary_sector']=master_frame['primary_sector'].str.replace('Na','0')
```

### Merging the master_frame data and Mapping data 

```python
master_frame=pd.merge(master_frame,mapping,how='inner',left_on='primary_sector',right_on='category_list')
master_frame = master_frame.drop(['category_list_y','Value'],axis=1)
```

### Check if there are any null values in the merged data

```python
master_frame.isnull().sum()
```

# Checkpoint 5.1

### Creating D1,D2,D3 for the top3 english speaking countries where the funding type is between 5million USD and 15 million USD

#### creation of D1 i.e., country as USA and funding type as venture and the investment amount between 5million and 15 million

```python
D1 = master_frame.loc[(master_frame.country_code=="USA") & (master_frame.funding_round_type =="venture") & (master_frame.raised_amount_usd >= 5000000)& (master_frame.raised_amount_usd <= 15000000)]
D1
```

#### Appending the `Total_amount_invested` and `Total_no.of_investments` columns for D1

```python
D1_sector=D1.groupby('main_sector').sum().rename(columns={'raised_amount_usd':'Total_amount_invested'})

D1_sector['Total_no.of_investments']=D1.main_sector.value_counts()
D1=pd.merge(D1,D1_sector,how='inner',left_on='main_sector',right_index=True)
```

```python
# Total number of Investments (in count)
D1.shape[0]
```

```python
# Total number of Investments (in USD)
D1.raised_amount_usd.sum()
```

### Sorting the D1 by `Total_no.of_investments` and find the top 3 invested sector and the companies with highest investment

```python
D1_sector.sort_values(by='Total_no.of_investments',ascending=False)
```

```python
D1.loc[D1.main_sector=='Social, Finance, Analytics, Advertising'].sort_values(by='raised_amount_usd',ascending=False)
```
